# Question 5: Number Pattern Generator

Create a program that prints numbers 1 to 100, but replaces multiples of 3 with "Boom", 
multiples of 5 with "Crash", and multiples of both with "BoomCrash". Allow users to 
customize the range, divisors, and replacement words.


```python
# Number Pattern Generator

start = int(input("Enter start of range: "))
end = int(input("Enter end of range: "))

div1 = int(input("Enter first divisor: "))
word1 = input("Enter word for first divisor: ")

div2 = int(input("Enter second divisor: "))
word2 = input("Enter word for second divisor: ")

for i in range(start, end + 1):
    if i % div1 == 0 and i % div2 == 0:
        print(word1 + word2)
    elif i % div1 == 0:
        print(word1)
    elif i % div2 == 0:
        print(word2)
    else:
        print(i)
```
