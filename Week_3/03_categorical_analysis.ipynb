## **Categorical Analysis:**

---

1. Compare average food waste by staff experience levels (bar plot)
2. Analyze waste distribution across different food categories (pie charts/bar plots)
3. Create seasonal waste analysis by food category (heatmap)
4. Examine the relationship between special events and meal volume

### **Import Data** 

```python
import pandas as pd 
df = pd.read_csv("processed_food_data.csv")
df.head()
```

## **1. Compare average food waste by staff experience levels (bar plot)**

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Compare average food waste by staff experience level using bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='staff_experience', y='past_waste_kg')
plt.title('Average Food Waste by Staff Experience Level')
plt.xlabel('Staff Experience Level')
plt.ylabel('Average Past Waste (kg)')
plt.show()
```

### **2. Analyze waste distribution across different food categories (pie charts)**

```python

import matplotlib.pyplot as plt

# Count each category
waste_category = df['waste_category'].value_counts()

# Plot pie chart
plt.pie(waste_category, labels=waste_category.index, autopct='%1.1f%%', startangle=90)
plt.title('Waste Category Distribution')
plt.show()
```

### **3. Create seasonal waste analysis by food category (heatmap)**

```python
import seaborn as sns
import matplotlib.pyplot as plt 

# Count waste occurrences by season and category
pivot_table = pd.crosstab(df['waste_category'], df['season'])

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title('Seasonal Waste Analysis by Food Category')
plt.show()
```

### 4. **Examine the relationship between special events and meal volume**

```python
plt.figure(figsize=(10,6))
sns.lineplot(
    data=df.assign(date=pd.to_datetime(df['date'])),
    x='date',
    y='past_waste_kg',
    hue='special_event',
    marker='o'
)
plt.title('Past Waste (kg) Trend over Time by Special Event')
plt.xlabel('Date')
plt.ylabel('Past Waste (kg)')
plt.xticks(rotation=45)
plt.legend(title='Special Event')
plt.tight_layout()
plt.show()
```
