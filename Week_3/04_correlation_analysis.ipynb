## **Import Data**

```python
import pandas as pd
df = pd.read_csv("processed_food_data.csv")
df.head()
```

```python
import seaborn as sns

# Correlation heatmap for key numerical variables
import matplotlib.pyplot as plt

cols = ['meals_served', 'kitchen_staff', 'temperature_C', 'humidity_percent', 'past_waste_kg']
num_df = df[cols].copy().apply(pd.to_numeric, errors='coerce')

corr = num_df.corr()

plt.figure(figsize=(7, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar_kws={"shrink": 0.8})
plt.title("Correlation Heatmap: meals, staff, temperature, humidity, past waste")
plt.tight_layout()
plt.show()
```

### **1. Plot correlation heatmap between numerical variables (meals_served, kitchen_staff,temperature, humidity, past_waste_kg)**

```python
import seaborn as sns

# Correlation heatmap for key numerical variables
import matplotlib.pyplot as plt

cols = ['meals_served', 'kitchen_staff', 'temperature_C', 'humidity_percent', 'past_waste_kg']
num_df = df[cols].copy().apply(pd.to_numeric, errors='coerce')

corr = num_df.corr()

plt.figure(figsize=(7, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar_kws={"shrink": 0.8})
plt.title("Correlation Heatmap: meals, staff, temperature, humidity, past waste")
plt.tight_layout()
plt.show()
```

## **2. Analyze relationship between weather conditions and food waste**

```python
import seaborn as sns
import matplotlib.pyplot as plt 

# Count waste occurrences by season and category
pivot_table = pd.crosstab(df['waste_category'], df['season'])

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title('Seasonal Waste Analysis by Food Category')
plt.show()
```

### **3. Study correlation between staff count and waste levels**

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Compare average food waste by staff experience level using bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='staff_experience', y='past_waste_kg')
plt.title('Average Food Waste by Staff Experience Level')
plt.xlabel('Staff Experience Level')
plt.ylabel('Average Past Waste (kg)')
plt.show()
```
