```python
import pandas as pd
df = pd.read_csv("processed_food_data.csv")
df.head()
```

### **1. Time series line plots showing meal trends and waste patterns over time**

```python
# Ensure 'date' is datetime and sorted for plotting
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.sort_values('date', inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Plot meal service trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='meals_served')
plt.title('Meal Service Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Meals Served')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Plot meal service trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='past_waste_kg')
plt.title('Meal watse Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Meals waste')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### **2. Seasonal heatmaps displaying waste by food category across seasons**

```python
import seaborn as sns
import matplotlib.pyplot as plt 

# Count waste occurrences by season and category
pivot_table = pd.crosstab(df['waste_category'], df['season'])

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title('Seasonal Waste Analysis by Food Category')
plt.show()
```

### **3. Bar plots comparing waste levels across staff experience categories**

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Compare average food waste by staff experience level using bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='staff_experience', y='past_waste_kg')
plt.title('Average Food Waste by Staff Experience Level')
plt.xlabel('Staff Experience Level')
plt.ylabel('Average Past Waste (kg)')
plt.show()
```

### **4. Box plots showing waste distribution by food categories**

```python
import seaborn as sns
import matplotlib.pyplot as plt 

sns.boxplot(data=df, x="waste_category", y="past_waste_kg")
plt.title("Box plots showing waste distribution by food categories")
```

### **5. Scatter plots analyzing relationships between meals served and waste generated**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x="meals_served", y="past_waste_kg", alpha=0.6)

plt.title("Scatter plot")
```

### **6. Interactive visualizations for exploring waste patterns by multiple dimensions**

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Plot meal service trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='past_waste_kg')
plt.title('Meal watse Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Meals waste')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
